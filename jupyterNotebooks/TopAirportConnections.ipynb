```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
df=pd.read_csv("static/files/AirportSummaryReduced.csv")
df
```

```python
df2=df.drop(['Distance', 'Cancelled'], axis=1)
```

```python
df2
```

```python
topAp = df2.groupby(by='Origin', as_index=False).agg({'Dest': pd.Series.nunique})
topAp
```

```python
topAp=topAp.sort_values(['Dest'], ascending=[False])
topAp=topAp.head(25)
topAp
```

```python
topApList = topAp['Origin'].tolist()
topApList
```

```python
df3 = df2[df2['Origin'].isin(topApList)]
```

```python
df3
```

```python
df4 = df3[df3['Dest'].isin(topApList)]
df4
```

```python
#df2["AverageDelay"] = ""
df4['AverageDelay'] = df4['DepDelay']# + df4['ArrDelay']
df4
```

```python
df5=df4.drop(['DepDelay', 'ArrDelay'], axis=1)
df5
```

```python
df6=df5.groupby(['Origin','Dest'])\
        .agg(AverageDelay=pd.NamedAgg(column='AverageDelay', aggfunc=lambda x: x.mean()))\
    .reset_index()
df6
```

```python
df6.to_csv("static/files/RouteDelay3.csv",index=False)
```
