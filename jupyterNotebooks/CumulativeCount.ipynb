```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
#fileName="MonthlyCSVs/AirPortCount"
fileName="MonthlyCSVs/AirLineCount"
```

```python
df=pd.read_csv(fileName+".csv", index_col=0)
df
```

```python
df2=df.cumsum(axis = 1)
df2
```
