```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
source_file="MonthlyCSVs/AirLineCountCumulavtive.csv"
```

```python
df_al=pd.read_csv("MonthlyCSVs/Airline_Lookup.csv")
df_al
```

```python
df_source=pd.read_csv(source_file)
df_source
```

```python
df_merge=pd.merge(df_al,df_source,on='IATA_Airline')
df_merge
```

```python
df_merge.drop('IATA_Airline', axis=1, inplace=True)
df_merge
```

```python
df_merge
```

```python
new_columns = df_merge.columns.values
new_columns[0] = 'IATA_Airline'
df_merge.columns = new_columns
df_merge
```

```python
df_merge = df_merge.sort_values(by=df_merge.columns[1], ascending=False)
df_merge
```

```python
df_merge.to_csv(source_file,index=False)
```
