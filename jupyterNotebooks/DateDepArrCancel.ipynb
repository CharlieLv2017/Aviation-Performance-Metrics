```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
df=pd.read_csv("static/files/ResultDataForAirports.csv")
df
```

```python
df2=df[['FlightDate','DepDelay','ArrDelay','Cancelled']]
df2
```

```python
#df2['DepDelay'] = pd.to_numeric(df2['DepDelay'])
```

```python
df_tot=df2.groupby('FlightDate').agg(Tot=pd.NamedAgg(column='FlightDate', aggfunc=lambda x: x.count())).reset_index()

minDelay=0

df_dep=df2[df2['DepDelay'] > minDelay].groupby('FlightDate')['DepDelay'].count().reset_index()
df_arr=df2[df2['ArrDelay'] > minDelay].groupby('FlightDate')['ArrDelay'].count().reset_index()
df_can=df2[df2['Cancelled'] > 0].groupby('FlightDate')['Cancelled'].count().reset_index()
```

```python
df_tot.iloc[0][1]
df_dep.iloc[1,1]
```

```python
for i in range(len(df_tot)):
    df_dep.iat[i,1]=(df_dep.iloc[i][1]*100)/df_tot.iloc[i][1]
    df_arr.iat[i,1]=(df_arr.iloc[i][1]*100)/df_tot.iloc[i][1]
    df_can.iat[i,1]=(df_can.iloc[i][1]*100)/df_tot.iloc[i][1]
```

```python
df_final=pd.merge(pd.merge(df_dep,df_arr,on='FlightDate'),df_can,on='FlightDate')
df_final
```

```python
#ap_df=createAirportCSV('MCI')
df_final.to_csv("static/files/csvData.csv",index=False)
```
