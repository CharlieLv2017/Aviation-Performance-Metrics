```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
df=pd.read_csv("static/files/AirportSummaryReduced.csv")
df
```

```python
df2=df
```

```python
df2=df2.groupby('Origin').filter(lambda x : len(x)>=50)
```

```python
df2=df2.groupby('Dest').filter(lambda x : len(x)>=50)
```

```python
#df2=df2[df2['Cancelled'] > 0]
df2
```

```python
df2.drop(df2.loc[df['Origin']=='EFD'].index, inplace=True)
df2.drop(df2.loc[df['Origin']=='ENV'].index, inplace=True)
df2.drop(df2.loc[df['Origin']=='IFP'].index, inplace=True)
df2.drop(df2.loc[df['Origin']=='TKI'].index, inplace=True)
df2.drop(df2.loc[df['Origin']=='YNG'].index, inplace=True)

df2.drop(df2.loc[df['Dest']=='EFD'].index, inplace=True)
df2.drop(df2.loc[df['Dest']=='ENV'].index, inplace=True)
df2.drop(df2.loc[df['Dest']=='IFP'].index, inplace=True)
df2.drop(df2.loc[df['Dest']=='TKI'].index, inplace=True)
df2.drop(df2.loc[df['Dest']=='YNG'].index, inplace=True)
```

```python
df_origin=df2.groupby('Origin')\
    .agg(AvgDist=pd.NamedAgg(column='Distance', aggfunc=lambda x: x.mean()),
        AvgDepDelay=pd.NamedAgg(column='DepDelay', aggfunc=lambda x: x.mean()))\
    .reset_index()
df_origin
```

```python
df_dest=df2.groupby('Dest')\
    .agg(AvgArrDelay=pd.NamedAgg(column='ArrDelay', aggfunc=lambda x: x.mean()))\
    .reset_index()
df_dest
```

```python
df_orig_tot=df2.groupby('Origin')\
    .agg(TotOrig=pd.NamedAgg(column='Origin', aggfunc=lambda x: x.count()))\
    .reset_index()
df_orig_tot
```

```python
df_dest_tot=df2.groupby('Dest')\
    .agg(TotDest=pd.NamedAgg(column='Dest', aggfunc=lambda x: x.count()))\
    .reset_index()
df_dest_tot
```

```python
df_onTimeDep=df2[df2['DepDelay'] <= 0].groupby('Origin')['DepDelay'].count().reset_index()
df_onTimeDep
```

```python
df_depCan=df2[df2['Cancelled'] > 0].groupby('Origin')['Cancelled'].count().reset_index()
df_depCan
```

```python
df_onTimeArr=df2[df2['ArrDelay'] <= 0].groupby('Dest')['ArrDelay'].count().reset_index()
df_onTimeArr
```

```python
for i in range(len(df_orig_tot)):
    df_onTimeDep.iat[i,1]=(df_onTimeDep.iloc[i][1]*100)/df_orig_tot.iloc[i][1]
    df_depCan.iat[i,1]=(df_depCan.iloc[i][1]*100)/df_orig_tot.iloc[i][1]
```

```python
for i in range(len(df_dest_tot)):
    df_onTimeArr.iat[i,1]=(df_onTimeArr.iloc[i][1]*100)/df_dest_tot.iloc[i][1]
```

```python
df_depCombi=pd.merge(pd.merge(df_origin,df_onTimeDep,on='Origin'),df_depCan,on='Origin')
df_depCombi
```

```python
df_arrCombi=pd.merge(df_dest,df_onTimeArr,on='Dest')
df_arrCombi
```

```python
df_depCombi = df_depCombi.rename(columns={'Origin': 'Airport', 'DepDelay': 'OnTimeDepRate', 'Cancelled': 'CancellationRate'})
df_depCombi
```

```python
df_arrCombi = df_arrCombi.rename(columns={'Dest': 'Airport', 'ArrDelay': 'OnTimeArrRate'})
df_arrCombi
```

```python
df_final=pd.merge(df_depCombi,df_arrCombi,on='Airport')
df_final
```

```python
df_final.to_csv("static/files/MapAirportSummary.csv",index=False)
```
